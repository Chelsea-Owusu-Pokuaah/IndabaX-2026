## 1. What is Machine Learning?

Traditional programming asks the developer to write the rules:

```
data + rules  ->  answers
```

Machine Learning flips this. We give the computer many examples of data paired with the right answers, and the algorithm finds the rules itself:

```
data + answers  ->  rules (model)
```

The "rules" the algorithm produces are called a **model**. Once trained, the model can be applied to new data the algorithm has never seen.

### A concrete example

Imagine sorting incoming emails into *spam* and *not-spam*.

- A rules-based programmer writes: *"if subject contains 'free money' then spam"*. This breaks the moment a spammer changes wording.
- An ML system is shown 10,000 labelled emails and learns which combinations of words, sender domains, and length tend to indicate spam. It generalises to new patterns the developer never anticipated.

Machine Learning is most useful when (a) the rules are too complex to enumerate by hand, and (b) we have enough labelled examples to learn from.


## 2. The three families of Machine Learning

| Family | What we give the algorithm | What it learns | Examples in a Ghanaian context |
|---|---|---|---|
| **Supervised learning** | Inputs **and** correct outputs (labels) | A mapping from inputs to outputs | Predicting cocoa pod ripeness from images; predicting whether a patient has heart disease from clinical features |
| **Unsupervised learning** | Inputs only, no labels | Hidden structure in the data | Segmenting Mobile Money users into spending personas; grouping crops by image similarity |
| **Reinforcement learning** | A simulated environment with rewards | A policy that maximises long-term reward | Optimising traffic-light timing at the Kwame Nkrumah Interchange; tuning energy distribution on a power grid |

This workshop focuses on **supervised classification** — predicting a category (e.g. *disease* vs *no disease*) from a fixed set of features.

## 3. The Machine Learning lifecycle

A real ML project is much more than calling `.fit()`. The lifecycle has roughly nine stages:

1. **Problem definition.** What decision is the model supporting? Whose life does it affect?
2. **Data collection.** Sourcing data ethically and with consent.
3. **Exploratory Data Analysis (EDA).** Understanding distributions, missingness, and obvious biases before modelling.
4. **Preprocessing.** Encoding categorical variables, scaling numeric ones, handling missing values.
5. **Train / validation / test split.** Setting aside data the model never sees during training, so we can honestly estimate how it will perform on new data.
6. **Model training.** Fitting an algorithm to the training set.
7. **Evaluation.** Measuring performance on held-out data using metrics appropriate to the problem (not just accuracy).
8. **Deployment.** Shipping the model to production so it can serve real predictions.
9. **Monitoring.** Watching for performance drift, data shifts, and fairness regressions over time.

Most beginner tutorials cover steps 4–7. The hardest, most consequential parts are 1, 2, and 9.


## 4. Core terminology

You will hear these terms repeatedly. Take a minute to understand each one.

- **Feature (input variable, X).** A column we use to make predictions. In the heart dataset we will use later, `age` and `cholesterol` are features.
- **Target (label, y).** The column we want to predict. In our case, `target` (1 = heart disease present, 0 = absent).
- **Sample (row, instance).** A single observation — one patient.
- **Training set.** The data the algorithm sees while learning (typically 70–80% of the data).
- **Test set.** Data held out and only used at the end to estimate real-world performance.
- **Model.** The function produced by training.
- **Parameters.** Values learned during training (e.g. coefficients in a linear model).
- **Hyperparameters.** Values *set by the developer* before training (e.g. number of trees in a random forest).
- **Generalisation.** How well the model performs on data it did not see during training.

### Overfitting and underfitting

A model **overfits** when it memorises quirks of the training data and fails on new data. Analogy: a student who memorises last year's exam questions word-for-word but cannot answer a slightly rephrased version this year.

A model **underfits** when it is too simple to capture the real pattern. Analogy: trying to draw a straight line through a curved trend.

The job of an ML practitioner is to find a model complex enough to capture the pattern but not so complex it memorises noise.

## 5. Garbage In, Garbage Out — and bias

A model can only be as fair as the data it is trained on.

Consider the **credit scoring dilemma**: a Ghanaian bank wants to use ML to approve loans. They train it on five years of historical approvals — but those approvals were dominated by salaried urban men in Accra. When the trained model is deployed and a Northern-region female shea-nut farmer applies, it has very few similar examples to learn from, and it is likely to reject her even if she is a strong credit risk. The model has *learned the bank's historical bias*.

### Three common sources of bias

- **Representation bias.** Some groups are under-represented in the training data. The model performs worse on them because it had less to learn from.
- **Measurement bias.** A feature is measured differently across groups (e.g. cholesterol tests done with different equipment).
- **Label bias.** The labels themselves encode historical unfairness (e.g. "approved for loan" reflects past discriminatory decisions, not actual creditworthiness).

### What to do about it

- Audit the dataset's demographics *before* training.
- Measure model performance *separately* on each subgroup, not just overall.
- Be willing to reject a more accurate model if it is systematically less fair.

## 6. Evaluating a classification model

Accuracy = (correct predictions) / (total predictions). It is the obvious metric and often the wrong one.

### Why accuracy is not enough

Suppose 1% of transactions are fraudulent. A model that always predicts *not fraud* will be 99% accurate — and useless. For imbalanced problems and high-stakes problems, we need richer metrics.

### The confusion matrix

For a binary classifier, every prediction falls into one of four cells:

|                       | Predicted: negative | Predicted: positive |
|---|---|---|
| **Actual: negative**  | True Negative (TN) | False Positive (FP) |
| **Actual: positive**  | False Negative (FN) | True Positive (TP) |

From this we derive:

- **Precision = TP / (TP + FP).** Of the cases the model flagged as positive, what fraction really were?
- **Recall = TP / (TP + FN).** Of the cases that were really positive, what fraction did the model catch?
- **F1 = 2 · (precision · recall) / (precision + recall).** A single number balancing the two.

### When to prefer precision vs recall

- Medical screening for a serious disease: prioritise **recall**. Missing a sick patient (false negative) is much worse than a false alarm.
- Spam filtering: prioritise **precision**. Misclassifying an important email as spam (false positive) is worse than letting one spam email through.

### ROC and AUC (preview)

A classifier usually outputs a probability, not a hard label. We turn the probability into a label by choosing a threshold (default 0.5). As we slide the threshold from 0 to 1, the trade-off between true-positive rate and false-positive rate traces out the **ROC curve**. The **AUC** (area under the curve) summarises the curve as a single number between 0.5 (random guessing) and 1.0 (perfect). We will plot one in Part 2.
